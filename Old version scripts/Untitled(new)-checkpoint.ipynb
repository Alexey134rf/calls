```python
import pandas as pd
import datetime as dt
import glob
import re

holidays = [dt.date(2020,12,31),dt.date(2021,1,1),dt.date(2021,1,2), dt.date(2021,1,3), dt.date(2021,1,4), dt.date(2021,1,5), 
            dt.date(2021,1,6), dt.date(2021,1,7), dt.date(2021,1,8), dt.date(2021,1,9), dt.date(2021,1,10), dt.date(2021,2,22),
            dt.date(2021,2,23), dt.date(2021,3,8), dt.date(2021,5,3), dt.date(2021,5,4), dt.date(2021,11,4), dt.date(2021,11,5)]

names1 = glob.glob('../Недозвоны/Ростелеком/*', recursive=True)
cont=[]
for i in range(len(names1)):
    date=re.search( r'[0-9]{4}-[0-9]{2}-[0-9]{2}', names1[i]).group()
    df = pd.read_excel(names1[i],dtype={'Номер вызывающего': str})
    if (dt.datetime.strptime(date, "%Y-%m-%d")<dt.datetime(2021,4,22)):
        df['Длительность']=df['Длительность'].str.replace( ' мин.', '').str.replace(' сек.', '')
        df['Длительность']=df['Длительность'].apply(lambda x: int(x.split(' ')[0])*60+int(x.split(' ')[1]) if len(x.split(' '))==2 else int(x.split(' ')[0]))
    else:
        df['Длительность']=pd.to_timedelta(df['Длительность'].astype('str'))
        df['Длительность']=df['Длительность'].apply(lambda x: x.total_seconds())
    df=df[df['Длительность'] > 9]
    
    df= df[['Дата вызова','Время вызова','Номер вызывающего','Номер вызываемого', 'Первый ответивший', 'Статус', 'Длительность', 'Группа']]
    df['Номер вызываемого']=df['Номер вызываемого'].astype('str')
    df['Время вызова']=df['Время вызова'].apply(lambda x: dt.datetime.strptime(x, '%H:%M:%S'))
    df['Дата вызова']=pd.to_datetime(df['Дата вызова'])
    df['День недели']=df['Дата вызова'].apply(lambda x: x.weekday())
    df['Час звонка']=df['Время вызова'].dt.hour
    df['Время вызова']=df['Время вызова'].dt.time
    if (pd.to_datetime(df['Дата вызова'].values[0]) in holidays):
        df=df[(df['Час звонка']>=8) & (df['Час звонка']<14)]
    else:
        df=df[(df['Час звонка']>=8) & (((df['День недели']<5) & (df['Час звонка']<18)) | ((df['День недели']>=5) & (df['Час звонка']<14)))]
    cont.append(df)
       
total_all_days = pd.concat(cont)
#total_all_days['Длительность']=total_all_days['Длительность'].apply(lambda x: x.total_seconds()/60)
total_all_days = total_all_days.rename(columns={'Первый ответивший': 'Оператор'})
total_all_days['Группа']=total_all_days['Группа'].replace({'Общая':'Медколледж','Консультация':'Облкомобразования'})
total_all_days['Группа']=total_all_days['Группа'].fillna('Не определена')
total_all_days['Оператор']=total_all_days['Оператор'].replace({' ':'Не определен'})
total_all_days['Номер вызывающего']=total_all_days['Номер вызывающего'].str.replace(' ', '')
total_all_days['Номер вызываемого']=total_all_days['Номер вызываемого'].str.replace(' ', '')
hospitals=pd.read_excel('Переадресация телефонов.xlsx')
total_all_days=total_all_days.merge(hospitals, how = 'left', left_on='Номер вызывающего', right_on='Единый номер')

total_all_days['Наименование учреждения'] = total_all_days['Наименование учреждения'].str.replace('Государственное учреждение здравоохранения','ГУЗ')
total_all_days['Наименование учреждения'] = total_all_days['Наименование учреждения'].str.replace('Государственное бюджетное учреждение здравоохранения','ГБУЗ') 
total_all_days['Наименование учреждения'] = total_all_days['Наименование учреждения'].str.replace('Клиническая больница скорой медицинской помощи','КБСМП')
total_all_days['Наименование учреждения'] = total_all_days['Наименование учреждения'].str.replace('клиническая поликлиника','КП') 
```

```python
import numpy as np 

def calculate_statistics2(dataframe):
    all_calls_count=dataframe.shape[0]
    all_calls_5552_count=dataframe[dataframe['Номер вызываемого']=='78442325552'].shape[0]
    all_calls_5554_count=dataframe[dataframe['Номер вызываемого']=='78442325554'].shape[0]
    missed_calls=dataframe[dataframe['Статус']=='пропущенный']
    missed_calls_count=missed_calls.shape[0]
    missed_calls_5552_count=missed_calls[missed_calls['Номер вызываемого']=='78442325552'].shape[0]
    missed_calls_5554_count=missed_calls[missed_calls['Номер вызываемого']=='78442325554'].shape[0]
    
    hospital_calls=dataframe[dataframe['Наименование учреждения'].isnull()==False]
    hospital_calls_count=hospital_calls.shape[0]
    hospital_calls_count_5552=hospital_calls[hospital_calls['Номер вызываемого']=='78442325552'].shape[0]
    hospital_calls_count_5554=hospital_calls[hospital_calls['Номер вызываемого']=='78442325554'].shape[0]
    print('Всего звонков: {0}, по линии 55-52: {1} ({3}%), по линии 55-54: {2} ({4}%)'.format(all_calls_count, all_calls_5552_count, all_calls_5554_count,np.round(all_calls_5552_count/ all_calls_count*100,2), np.round(all_calls_5554_count/ all_calls_count*100,2) ))
    print('Неотвеченных вызовов: {0}, по линии 55-52: {1} ({3}%), по линии 55-54: {2} ({4}%)'.format(missed_calls_count,missed_calls_5552_count, missed_calls_5554_count,np.round(missed_calls_5552_count/ missed_calls_count*100,2), np.round(missed_calls_5554_count/ missed_calls_count*100,2) ))
    print('Процент неотвеченных вызовов: {0} %'.format(np.round(missed_calls_5552_count/all_calls_5552_count*100,2)))
    print('Переадресовано из мед.учреждений всего: {0} ({3}% от вызовов на 55-54), по линии 55-52: {1}, по линии 55-54: {2}'.format(hospital_calls_count, hospital_calls_count_5552, hospital_calls_count_5554, np.round(hospital_calls_count/all_calls_5554_count*100)))
```
